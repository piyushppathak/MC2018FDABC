```python
import pandas as pd

loc = 'GleeClubData.xlsx'
df = pd.read_excel(loc)

dfNew = df.copy()
dfNew = dfNew.rename(index=str, columns={'team size' : 'TeamSize', 'male/female ratio of team' : 'TeamMFRatio', 'enrollment in school' : 'SchoolEnrollNum', 'male/female ratio of School' : 'SchoolMFRatio', 'population of district served by school' : 'DistPopuSrvdbySch', '# of practices per month' : 'NumPractPerMonth', '# of competitions before regionals' : 'NumCompBeforRegional' ,'Regionals Score' : 'RegScor'})

#dfNew.head()

dfNew.corr()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='RegScor ~ TeamSize + TeamMFRatio + SchoolEnrollNum + SchoolMFRatio + DistPopuSrvdbySch + NumPractPerMonth + NumCompBeforRegional - 1', data=dfNew).fit()
result.summary()
```
