```python
## Load titanic3.xls

import pandas as pd

fileloc = "titanic3.xls"
df = pd.read_excel(fileloc)
df.head()
```

```python
def gender_to_numeric(gender):
    if gender == 'female':
        return 1
    if gender == 'male':
        return 0
```

```python
df['sex'] = df['sex'].apply(gender_to_numeric)
df.head()
```

```python
def get_family_name(fullname):
    return fullname.split(',')[0]
```

```python
df['family'] = df['name'].apply(get_family_name)
df.head()
```

```python
import numpy as np

df['relatives'] = np.nan
df.head()
```
